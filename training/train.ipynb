```python
from ultralytics import YOLO

# ëª¨ë¸ ë¶ˆëŸ¬ì˜¤ê¸° (ê°€ìž¥ ê°€ë²¼ìš´ YOLOv8 ëª¨ë¸)
model = YOLO("yolov8n.pt")

# í›ˆë ¨ ì‹œìž‘
model.train(
    data="/home/workspace/dbd/dataset/data.yaml",  # ë°ì´í„°ì…‹ YAML ê²½ë¡œ
    epochs=60,
    imgsz=640,
    batch=32,
    name="cigarette-detector-final",
    device=1  # GPU 1ë²ˆ ì‚¬ìš©
)
```

```python
import matplotlib.pyplot as plt

model = YOLO('/home/workspace/dbd/runs/detect/cigarette-detector-final/weights/best.pt')
results = model.val(data='/home/workspace/dbd/dataset/data.yaml')

print(f"mAP50: {results.box.map50:.4f}")
print(f"mAP50-95: {results.box.map:.4f}")

# ì‹œê°í™” ê·¸ëž˜í”„ ë³´ê¸°
base = '/home/workspace/dbd/runs/detect/cigarette-detector-final/val/'
for imgfile in ['confusion_matrix.png', 'pr_curve.png', 'F1_curve.png']:
    plt.figure(figsize=(6,6))
    plt.imshow(plt.imread(base + imgfile))
    plt.axis('off')
    plt.title(imgfile.replace('.png','').replace('_',' ').title())
    plt.show()
```
